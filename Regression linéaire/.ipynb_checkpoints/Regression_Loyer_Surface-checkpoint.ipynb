```python
# On importe les librairies dont on aura besoin pour ce tp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

------
<a id="ref1"></a> Données

```python
# charge le dataset
house_data = pd.read_csv('house.csv')

# affiche le nuage de points dont on dispose
plt.plot(house_data['surface'], house_data['loyer'], 'ro', markersize=4)
plt.show()
```

-------
<a id="ref2"></a> Nettoyage Données

```python
house_data=house_data[house_data['loyer']<7000]

plt.plot(house_data['surface'], house_data['loyer'], 'ro', markersize=4)
plt.show()
```

------
<a id="ref3"></a> Régression l'inéaire

X : Matrice comportant un vecteur de 1 + donnees surface
    
Y : montant du loyer

```python
X = np.matrix([np.ones(house_data.shape[0]),house_data.surface]).T
X
```

```python
Y = np.matrix(house_data.loyer).T
Y[0:10]
```

```python
# On effectue le calcul exact du paramètre theta
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)

print(theta)
```

---------
<a id="ref4"></a> Modèle finale

```python

plt.xlabel('Surface')
plt.ylabel('Loyer')

plt.plot(house_data['surface'], house_data['loyer'], 'ro', markersize=4)

# On affiche la droite entre 0 et 250
plt.plot([0,250], [theta.item(0),theta.item(0) + 250 * theta.item(1)], linestyle='--', c='#000000')

plt.show()
```

---------
<a id="ref5"></a> Prédictions pour un loyer de 50 m^2

```python
theta.item(0) + theta.item(1) * 50
```
